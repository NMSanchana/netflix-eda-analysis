```python
df["cast"].isnull().sum()
```

```python
df["rating"].fillna("Not rated",inplace=True)
```

```python
df["rating"].isnull().sum()
```

```python
df["duration"].fillna("Avg duration",inplace=True)
```

```python
df["duration"].isnull().sum()
```

```python
df.dropna(subset=["date_added"],inplace=True)   #date_added null values can be removed since we don't want to distort analysis
```

```python
## 7 Content Type Distribution
df["type"].value_counts()
```

```python
sns.countplot(data=df,x="type",color="pink")
plt.title("Types of Show")
plt.show()
```

```python
## 8 Content Release Trend Over Years
year_counts = df["release_year"].value_counts()
```

```python
year_counts = year_counts.sort_index()
```

```python
plt.figure(figsize=(10,5))
plt.plot(year_counts.index, year_counts.values,color="pink")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.title("Netflix Content Release Over Years")
plt.show()
```

```python
## 9 Top Countries Producing Netflix Content
df["country"].value_counts().head(10)
```

```python
top_countries = df["country"].value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.values, y=top_countries.index,color="pink",edgecolor="black")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.title("Top 10 Countries Producing Netflix Content")
plt.show()
```

```python
## 10 Most Popular Genres on Netflix
df["main_genre"] = df["listed_in"].str.split(",").str[0]
```

```python
v_c = df["main_genre"].value_counts().head(10)
v_c
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=v_c.values,y=v_c.index,color="pink",edgecolor="black")
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Number of titles")
plt.ylabel("Genre")
plt.show()
```

```python
## 12 Distribution of Release Years
plt.figure(figsize=(10,5))
sns.histplot(x="release_year",data=df,bins=20,color="pink",edgecolor="black")
plt.title("Distribution of Release Years")
plt.xlabel("Release_year")
plt.ylabel("Count")
plt.yticks(np.arange(0, 5001, 500))
plt.show()
```

```python
## 13 Movie Duration Distribution
movies = df[df["type"] == "Movie"]
```

```python
movies.dropna(inplace=True)
```

```python
movies["duration_int"] = movies["duration"].str.replace(" min","").astype(int)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(movies["duration_int"], bins=30, color="pink", edgecolor="black", kde=True)
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")
plt.title("Distribution of Movie Durations")
plt.show()
```

```python
## 14 Movie Duration Boxplot
plt.figure(figsize=(10,5))
sns.boxplot(x=movies["duration_int"],color="pink")
plt.xlabel("Movie Duration (minutes)")
plt.title("Boxplot of Movie Durations")
plt.show()
```

```python
## 15 Violin Plot of Movie Durations
plt.figure(figsize=(10,5))
sns.violinplot(y=movies["duration_int"],color="pink")
plt.title("Violin Plot of Movie Durations")
plt.show()
```

```python
## 16 Release Year vs Movie Duration
plt.figure(figsize=(10,5))
sns.scatterplot(x=movies["release_year"], y=movies["duration_int"], s=30, color="pink")
plt.xlabel("Release Year")
plt.ylabel("Duration (minutes)")
plt.title("Release Year vs Movie Duration")
plt.show()
```

```python
## 17 Multiple Visualizations
fig, ax = plt.subplots(1,2, figsize=(12,5))

sns.countplot(x="type", data=df, ax=ax[0], color="pink")
ax[0].set_title("Content Type Distribution")

sns.histplot(df["release_year"], bins=30, ax=ax[1], color="pink", edgecolor="black")
ax[1].set_title("Release Year Distribution")

plt.show()
```

```python
## 18 Pairplot Analysis
pair_data = movies[["release_year","duration_int"]]
sns.pairplot(pair_data)
plt.show()
```

```python
## 19 Correlation Heatmap
corr = pair_data.corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
